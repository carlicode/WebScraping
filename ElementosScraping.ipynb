**Código de respuesta HTTP**

Códigos de estado de respuesta HTTP
Los códigos de estado de respuesta HTTP indican si se ha completado satisfactoriamente una solicitud HTTP específica. Las respuestas se agrupan en cinco clases:


1.   Respuestas informativas (100–199)
2.   Respuestas satisfactorias (200–299)
3.   Redirecciones (300–399)
4.   Errores de los clientes (400–499)
5.   y errores de los servidores (500–599)



```python
import requests
import csv
from bs4 import BeautifulSoup

def estadoHTTP(url):
    response = requests.get(url)
    print(response.status_code)


url_tabla_periodica = "https://es.wikipedia.org/wiki/Tabla_peri%C3%B3dica_de_los_elementos"
estadoHTTP(url_tabla_periodica)
```

**Impresión de texto plano**

```python
def estadoHTML(url):
    response = requests.get(url)
    soup = response.text
    print(soup)

url_tabla_periodica = "https://es.wikipedia.org/wiki/Tabla_peri%C3%B3dica_de_los_elementos"
estadoHTML(url_tabla_periodica)
```

# Información de imágenes, audios, etc

```python
def contenido(url):
    response = requests.get(url)
    soup = response.content
    print(soup)


url_tabla_periodica = "https://es.wikipedia.org/wiki/Tabla_peri%C3%B3dica_de_los_elementos"
contenido(url_tabla_periodica)
```

# Información sobre fecha, tipo de contenido, info sobre el servidor

```python
def encabezado(url):
    response = requests.get(url)
    soup = response.headers
    print(soup)


url_tabla_periodica = "https://es.wikipedia.org/wiki/Tabla_peri%C3%B3dica_de_los_elementos"
encabezado(url_tabla_periodica)
```

```python
def encabezadoSolicitud(url):
    response = requests.get(url)
    soup = response.request.headers
    print(soup)

url_tabla_periodica = "https://es.wikipedia.org/wiki/Tabla_peri%C3%B3dica_de_los_elementos"
encabezadoSolicitud(url_tabla_periodica)
```

# Sopa con identación y jerarquía HTML

```python
def encabezadoSolicitud(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "lxml")
  print(soup.prettify())

url_tabla_periodica = "https://es.wikipedia.org/wiki/Tabla_peri%C3%B3dica_de_los_elementos"
encabezadoSolicitud(url_tabla_periodica)
```

# Lista de items de Elementos de transición



```python
def encabezadoSolicitud(url_tabla_periodica):
  response = requests.get(url_tabla_periodica)
  soup = BeautifulSoup(response.text, "lxml")
  table = soup.find('table', attrs={
      'class' : 'wikitable col2cen col3cen col4cen col5cen col6cen col7cen col8cen col9cen'}).find_all('td')
  print(table)

url_tabla_periodica = "https://es.wikipedia.org/wiki/Tabla_peri%C3%B3dica_de_los_elementos"
encabezadoSolicitud(url_tabla_periodica)
```
